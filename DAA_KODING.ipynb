```python
import pandas as pd
import re

# Baca data
data = pd.read_csv('/content/drive/MyDrive/DAA/data_test.csv', encoding='latin-1', header=0, names=['id', 'job_title', 'location', 'career_level','experience_level', 'education_level', 'salary'])
df = pd.DataFrame(data)

# Menghapus baris yang memiliki nilai NaN pada kolom 'salary'
df = df.dropna(subset=['salary'])

# Mengurutkan DataFrame berdasarkan gaji tertinggi
df_sorted_by_salary = df.sort_values(by='salary', ascending=False)

# Data User
user_job = input('Pekerjaan yang ingin dicari: ')
user_location = input('Lokasi anda: ')

# Linear Search untuk multiple hasil
def linear_search_multiple(dataframe, job, location):
    results = []
    for index, row in dataframe.iterrows():
        if re.search(job, row['job_title'], flags=re.IGNORECASE) and re.search(location, row['location'], flags=re.IGNORECASE):
            results.append(row)
    return results

# Menerapkan Linear Search untuk multiple hasil berdasarkan input user
results = linear_search_multiple(df_sorted_by_salary, user_job, user_location)

# Menampilkan hasil pencarian
if results:
    print("\nHasil Pencarian:")
    for result in results:
        print(result[['job_title', 'location', 'experience_level', 'education_level', 'salary']])
else:
    print('\nPekerjaan tidak ditemukan.')
```

```python
import pandas as pd
import re
```

```python
# Baca data
data = pd.read_csv('/content/drive/MyDrive/DAA/data_test.csv', encoding='latin-1', header=0, names=['id', 'job_title', 'location', 'career_level','experience_level', 'education_level', 'salary'])
df = pd.DataFrame(data)
data.head(30)
```

```python
# Menghapus baris yang memiliki nilai NaN/kosong pada setiap kolom
df = df.dropna(subset=['job_title', 'location', 'career_level','experience_level', 'education_level', 'salary'])
df.head(30)
```

```python
# Mengurutkan DataFrame berdasarkan gaji tertinggi
df_sorted_by_salary = df.sort_values(by='salary', ascending=False)
df.head(20)
```

```python
# membuat fungsi untuk melakukan pencarian
def linear_search_multiple(dataframe, job, location, education):
    results = []
    for index, row in dataframe.iterrows():
        if re.search(job, row['job_title'], flags=re.IGNORECASE) and re.search(location, row['location'], flags=re.IGNORECASE) and re.search(education, row['education_level'], flags=re.IGNORECASE):
            results.append(row)
    return results
```

```python
# Data User
user_job = input('Pekerjaan yang ingin dicari: ')
user_location = input('Lokasi anda: ')
user_education = input('Pendidikan terakhir anda: ')

# Menerapkan Linear Search untuk multiple hasil berdasarkan input user
results = linear_search_multiple(df_sorted_by_salary, user_job, user_location,user_education)

# Menampilkan hasil pencarian
if results:
    print("\nHasil Pencarian:")
    for result in results:
        print(result[['job_title', 'location', 'experience_level', 'education_level', 'salary']])
else:
    print('\nPekerjaan tidak ditemukan.')
```
